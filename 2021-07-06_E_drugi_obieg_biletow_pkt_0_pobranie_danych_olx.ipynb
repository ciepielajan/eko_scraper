__pobranie danych z każdej oferty__

```python
#0 nazwy_plików
csv_linki_do_ofert = '2021-07-01_olx.csv'
csv_linki_do_ofert_columna = "link"

plik_z_pobranymi_ofertami = '2021-07-01_olx_ogloszenia.csv'
plik_z_pobranymi_ofertami_columny = ["offer_id", "title", "opis", "link", "price", 
                                     "offer_date", "offer_date_end", "city", "breadcrumb", "views", 
                                     "data_wydarzenia", "miejsce_wydarzenia", "wersja", "params"]

path_to_site_view = "scrap_site_view/olx/2021-07-09_"


"2021-07-01_olx_ogloszenia.csv"

#1 pobranie wszystkich linków do pobrania
df_wszystkie = pd.read_csv(csv_linki_do_ofert, usecols=[csv_linki_do_ofert_columna])

#2 pobranie już pobranych linków, jeżeli brak to utwórz pusty plik pod pierwsze pobieranie
try:
    df_offers = pd.read_csv(plik_z_pobranymi_ofertami)
except:
    df_offers = pd.DataFrame(list(), columns=plik_z_pobranymi_ofertami_columny)
    df_offers.to_csv(plik_z_pobranymi_ofertami, index=False)

#3 utworzenie listy brakujących linków
unikalne_klucze_left = df_wszystkie.link.to_list()
unikalne_klucze_right = df_offers.link.to_list()

missing_offers = [i for i in unikalne_klucze_left if i not in unikalne_klucze_right]


#4 zmienne od funkcji liczącej postęp
number_of_values_left = len(missing_offers)
timeout_list = []
start = time.time()


#5 rozpoczęcie pobierania
print(time.ctime() )
for idx, offer in enumerate(missing_offers, start=1):
    sleep(3)
    print(f"LINK: {offer}")
    user_agent = random.choice(user_agent_list)
    headers = {'User-Agent': user_agent}
    
    
    ave_time = (time.time() - start) / idx
    left_litr = number_of_values_left - idx
    left_litr_time = time.strftime('%H:%M:%S', time.gmtime(left_litr * ave_time))
    print(idx, "z", number_of_values_left, "pozostało:", left_litr, "przewidywany czas: ", left_litr_time)
    
    
    
    try:
        page = get(offer, headers=headers, timeout =10 ,verify=False)   
        print(page)
        print()

        bs = BeautifulSoup(page.content, "html.parser")

        link = offer 
        
        try:
            offer_id = bs.find("span", class_="css-9xy3gn-Text").get_text().strip()
        except:
            offer_id = ""

        try:
            title = bs.find("h1", class_="css-1oarkq2-Text").get_text().strip()
        except:
            title = ""


        try:
            opis = bs.find("div", class_="css-g5mtbi-Text").get_text().strip()
        except:
            opis = ""



        try:
            price = bs.find("h3", class_="css-8kqr5l-Text").get_text().strip()
        except:
            price = ""


        try:
            offer_date = bs.find("span", class_="css-19yf5ek").get_text().strip()
        except:
            offer_date = ""

        try:
            offer_date_end = ""
        except:
            offer_date_end = ""

        try:
            city = ""
        except:
            city = ""


        views = ""

        try:
            breadcrumb = " / ".join([i.get_text().strip() for i in bs.find("ol", class_="css-2tdfce").find_all("li")])
        except:
            breadcrumb = ""



        try:
            params = ""
        except:
            params = ""


        try:
            data_wydarzenia = ""
        except:   
            data_wydarzenia = ""
            
            
        try:
            miejsce_wydarzenia = ""
        except:   
            miejsce_wydarzenia = ""

            
        try:
            wersja = ""
        except:   
            wersja = ""


        values = {
            "offer_id": offer_id,
            "title":title,
            "opis":opis,
            "link":link,
            "price":price,
            "offer_date":offer_date,
            "offer_date_end": offer_date_end,
            "city":city,
            "breadcrumb":breadcrumb,
            "views":views,
            "data_wydarzenia":data_wydarzenia,
            "miejsce_wydarzenia":miejsce_wydarzenia,
            "wersja":wersja,
            "params":params,
        }

#         for key in values:
#             print(key,":", values[key])


        try:
            name_file = offer.split("/")[-1].split(".html")[0]
            file1 = open(f'{path_to_site_view}{name_file}.txt', 'w+')
            file1.write(str(page.text.encode('utf-8')))
            file1.close()
            print("plik z obrazem strony zapisany poprawnie")
        except:
            print("BŁĄD ZAPISU: plik z obrazem strony")
            

#         row_offers = [offer_id, title, opis, link, price, offer_date, offer_date_end, city, breadcrumb, views, data_wydarzenia, miejsce_wydarzenia, wersja, params]
        df_offers = df_offers.append(values, ignore_index = True)
        
        try:
#         if (idx % 100==0):
            df_offers.to_csv(plik_z_pobranymi_ofertami, index=False)
            print("plik z CSV zapisany poprawnie")
        except:
            print("BŁĄD ZAPISU: plik CSV")
        
        
    except requests.exceptions.Timeout:
        timeout_list.append(offer)
        print("\n\n\n\n\n Timeout occurred \n\n\n\n\n ")
        
        
    print("\n--------------------------------------\n")

print("--THE END--")
print(time.ctime() )
```
