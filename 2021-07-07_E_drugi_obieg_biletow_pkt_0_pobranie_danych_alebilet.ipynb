__Schemat działania:__
* 1. Pobranie linków do artystów ze poszczególnych stron kategorii
* 2. Wprawdzenie czy link przekierowywuje bezpośrednio do strony wydarzenia (w przypadku tylko jednego wydarzenia danego artysty), czy do strony do wielu wydarzeń:
* 2.1. Jedno wydarzenie:
* 2.1.1.  ściągnij informację danego wydarzenia
* 2.2. Wiele wydarzeń:
* 2.2.1. Pobierz listę wydarzeń
* 2.2.2. ściągnij infomrację danego wydarzenia
* 3.1. Pobierz informacje o wszytkich biletach

```python
# Run the argument with incognito
# chrome_options.add_argument("--incognito") # używać bez VNP


list_website = [
    "https://www.alebilet.pl/festiwale" ,
    "https://www.alebilet.pl/koncerty" ,
    "https://www.alebilet.pl/sport" ,
    "https://www.alebilet.pl/scena" 
]

list_links_to_artists = list()

# Getting the start time to track on time required
start = time.time()

driver = webdriver.Chrome(executable_path='./selenium/chromedriver', chrome_options=chrome_options)


for website in list_website:
# -------------------------------Web Scraping-------------------------------
    driver.get(website) 

    print(f"Start: {time.ctime()}")
    print(f"LINK: {website}")

    # Wait 30 seconds for page to load and extract the element after it loads
    timeout = 30
    try:
        WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.ID, "category-page")))
    except TimeoutException:
        print('Timed out waiting for page to load')
        driver.quit()

    # -------------------------------links to artists-------------------------------
    table = driver.find_element_by_class_name('events-wrapper')
    rows = table.find_elements_by_class_name('event-item')
    for row in rows:
        row = row.get_attribute("onclick")
        link = row.strip().split('"')[-2]
        link = "https://www.alebilet.pl" + link
        print(link)
        list_links_to_artists.append(link)


    
    
    
```

```python
#0 nazwy_plików
csv_linki_do_ofert = '2021-07-12_alebilet.csv'

plik_z_pobranymi_ofertami = '2021-07-12_alebilet_ogloszenia-21_07.csv'
plik_z_pobranymi_ofertami_columny = ["date_scrap",
                                        "breadcrumb",
                                        "title",
                                        "desc",
                                        "location",
                                        "event_date",
                                        "event_time" ,
                                        "brak_biletow",
                                        "area",
                                        "quantity",
                                        "availability",
                                        "price"]



path_to_site_view = "scrap_site_view/alebilet_2/2021-07-12_"
```

```python
def scrap_tickets(driver):
    date_scrap =  time.strftime("%d-%m-%Y, %H:%M")
    
    try:
        breadcrumb = [i.text.strip() for i in driver.find_elements_by_css_selector("ol.breadcrumb > li ")][1]
    except:
        breadcrumb = ""
    print("breadcrumb: ",breadcrumb)
    
    try:
        title = driver.find_element_by_css_selector("div.event-desc > h1 ").text.split("Bilety na")[1].strip()
    except:
        title = ""
    print("Title: ",title)
    
    try:
        desc = driver.find_element_by_css_selector("div.event-desc > h4").text.strip()
    except:
        desc = ""
    print("desc: ",desc)
    
    try:
        location = driver.find_element_by_css_selector("div.event-location > a").text.strip()
    except:
        location = ""
    print("location: ",location)
    
    try:
        event_date = driver.find_element_by_css_selector("div.event-date").text.strip()
    except:
        event_date = ""
    print("event-date: ",event_date)
    
    try:
        event_time = driver.find_element_by_css_selector("div.event-time").text.strip()
    except:
        event_time = ""
    print("event-time: ",event_time)
    
    try:
        tickets = driver.find_elements_by_class_name('category')
        
        
        if len(tickets) == 0:     
            print("brak biletw")
            brak_biletow = True
            tickets.append("")
        else:
            brak_biletow = False
    
        for row in tickets:

            try:
                area = row.find_element_by_css_selector("span.cat").text.strip()
            except:
                area = ""

            print("Area",area)    

            try:
                quantity = row.get_attribute("data-quantity")
            except:
                quantity = ""
            print("Quantity",quantity)

            try:
                availability = row.find_element_by_css_selector("td.availability").text.strip()
            except:
                availability = ""
            print("availability",availability)

            try:
                price = row.find_element_by_css_selector("td.price").text.replace("\n", " ").strip()
            except:
                price = ""
            print("price",price)
    except:
        print("brak biletw")
    print("\n\n\n--------------------------")
    
    
    result = {
        "date_scrap":date_scrap,
        "breadcrumb":breadcrumb,
        "title":title,
        "desc":desc,
        "location":location,
        "event_date":event_date,
        "event_time":event_time ,
        "brak_biletow": brak_biletow,
        "area":area,
        "quantity":quantity,
        "availability":availability,
        "price":price,
    }
    
    return result
                             
```

```python
# Getting the start time to track on time required
start = time.time()

driver = webdriver.Chrome(executable_path='./selenium/chromedriver', chrome_options=chrome_options)

# Getting the start time to track on time required
start = time.time()

#1 pobranie wszystkich linków do pobrania
df_wszystkie = pd.read_csv(csv_linki_do_ofert)
wszystkie_linki = df_wszystkie["0"].to_list()

#2 pobranie już pobranych linków, jeżeli brak to utwórz pusty plik pod pierwsze pobieranie
try:
    df_offers = pd.read_csv(plik_z_pobranymi_ofertami)
except:
    df_offers = pd.DataFrame(list(), columns=plik_z_pobranymi_ofertami_columny)
    df_offers.to_csv(plik_z_pobranymi_ofertami, index=False)


#4 zmienne od funkcji liczącej postęp
number_of_values_left = len(df_wszystkie)
timeout_list = []
start = time.time()
print(time.ctime() )

#5 rozpoczęcie pobierania
    

# for website in unique_list_links_to_artists[:20]:
# for website in in enumerate(missing_offers, start=1):
# df_wszystkie.values
for idx, website in enumerate(wszystkie_linki, start=1):
# -------------------------------Web Scraping-------------------------------
    driver.get(website) 

    print(f"Start: {time.ctime()}")
    print(f"LINK: {website}")
    ave_time = (time.time() - start) / idx
    left_litr = number_of_values_left - idx
    left_litr_time = time.strftime('%H:%M:%S', time.gmtime(left_litr * ave_time))
    print(idx, "z", number_of_values_left, "pozostało artystów:", left_litr, "przewidywany czas zakończenia: ", left_litr_time)
    
    
    

    # Wait 30 seconds for page to load and extract the element after it loads
    timeout = 30
    try:
        WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.ID, "container")))
    except TimeoutException:
        print('Timed out waiting for page to load')
        driver.quit()

# --- Pobranie linków do wydarzeń artysty:
    try:
        table_events = driver.find_element_by_class_name('events-wrapper')
    except:
        table_events = ""
        
        
    try:
        table_tickets = driver.find_element_by_id('ticket-list')
    except:
        table_tickets = ""
        
    
    
    list_links_to_events = list()
    
    if (table_events == "") & (table_tickets == ""):
        print("BRAK: table_events i table_tickets")
        break
        
    
    elif (table_events != "") & (table_tickets == ""):
        print("events events events events events events events")
        print("Pobieram linki do wydarzeń")
        

        rows = driver.find_elements_by_css_selector('tr.event-item')

        for row in rows:
            row = row.get_attribute("onclick")
            link = row.strip().split("'")[-2]
            link = "https://www.alebilet.pl" + link
            print(link)
            list_links_to_events.append(link)
            
            
            
    elif (table_events == "") & (table_tickets != ""):
        print("tickets tickets tickets tickets tickets ")
        print("Pobieram wydarzenie") 
        
        list_links_to_events.append(website)
            
            
    # wejź na każdli link i pobierz dane bilety wydarzenia
    for website_event in list_links_to_events:
        driver.get(website_event) 

        print(f"Start: {time.ctime()}")
        print(f"LINK: {website_event}")

        # Wait 30 seconds for page to load and extract the element after it loads
        timeout = 30
        try:
            WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.ID, "container")))
        except TimeoutException:
            print('Timed out waiting for page to load')
            driver.quit()

        values = scrap_tickets(driver)


        # zapis obrazu strony
        try:
            name_file = "_".join(website_event.split("/")[4:]).replace(":","")
            print(name_file)
            print(path_to_site_view)
            file1 = open(f'{path_to_site_view}{name_file}.txt', 'w+')
            file1.write(str((driver.page_source).encode('utf-8')))
            file1.close()
            print("plik z obrazem strony zapisany poprawnie 1")
        except:
            print("BŁĄD ZAPISU: plik z obrazem strony 1")

        # zapis danych do pliku csv

        df_offers = df_offers.append(values, ignore_index = True)

        try:
            df_offers.to_csv(plik_z_pobranymi_ofertami, index=False)
            print("plik z CSV zapisany poprawnie 1")
        except:
            print("BŁĄD ZAPISU: plik CSV 1")
   

        


print("--THE END--")
print(time.ctime() )
```
