# Unormowanie cech

## cechy datetime

```python
import re
```

```python
month2nub = {'cze': '06',
 'gru': '12',
 'kwi': '04',
 'lip': '07',
 'lis': '11',
 'lut': '02',
 'maj': '05',
 'mar': '03',
 'paź': '10',
 'sie': '08',
 'sty': '01',
 'wrz': '09'}
```

```python
def str2data(i):
    i = str(i).replace("/", ".")
    i = i.split(" ")
#     print(i)
    for idx, x in enumerate(i):
        for y in month2nub.keys():
            if y in x:
#                 print(f"znalzałem ----> {month2nub[y]}")
#                 print(f"{y} dla {x} zamienie na: {month2nub[y]}")
                i[idx] = str(month2nub[y])
            else:
                continue
    
    i = " ".join(i)
    i = i.split(",")
    i = i[0].replace(" ",".")
    
    ## wyłąpane błędy
    i = i.replace("26.10.19","26.10.2019")
    i = re.sub("\.+",".", i)
    
#     print(i)
#     print("\n")
    return i


scrap_all_offers["event_data_clean"] = scrap_all_offers["event_data"].apply(lambda x: str2data(x))
```

```python
# data =  scrap_all_offers[scrap_all_offers["event_data_clean"] != "nan"]
data =  scrap_all_offers["event_data_clean"]

scrap_all_offers["event_data_clean"] = pd.to_datetime(data, format='%d.%m.%Y', utc=True)
```

```python
data = scrap_all_offers["offer_date_start"].str.extract(r"([0-9]+\-[0-9]+\-[0-9]+)", expand=False)

scrap_all_offers["offer_date_start_clean"] = pd.to_datetime(data, format='%d-%m-%Y', utc=True)
```

```python
data = scrap_all_offers["offer_date_end"].str.extract(r"([0-9]+\.[0-9]+\.[0-9]+)", expand=False)

scrap_all_offers["offer_date_end_clean"] = pd.to_datetime(data, utc=True)
```

```python
data =  scrap_all_offers["date_scrap"]

scrap_all_offers["date_scrap_clean"] = pd.to_datetime(data, format='%Y-%m-%d', utc=True)
```

__Wybór jednego terminu dla wydarzenia__

```python
def min_value(date):
    try:
        return min([ x for x in date if pd.notnull(x)])
    except:
        return 0

scrap_all_offers["match_data_min"] = scrap_all_offers[["event_data_clean","offer_date_start_clean","offer_date_end_clean"]].apply(lambda x: min_value(x), axis=1)
```
